# Taxi Dataset from AWS S3 using Jupyter notebook

```python
import boto3
import pandas as pd
import pandas_profiling
```

```python
client = boto3.client('s3')
```

```python
bucket = "yellowtaxibucket"
file_name = "yellow_tripdata_2020-12.csv"

s3 = boto3.client('s3') 
# 's3' is a key word. create connection to S3 using default config and all buckets within S3

obj = s3.get_object(Bucket= bucket, Key= file_name) 
# get object and file (key) from bucket

df = pd.read_csv(obj['Body'], low_memory=False) # 'Body' is a key word, low_memory - guessing dtypes for each column is very memory demanding

df.head()
```

```python
df.head(100)
```

```python
profile = pandas_profiling.ProfileReport(df)
print(profile)
```

## Profile dataset using pandas

```python
# run the profile report
profile = df.profile_report(title='Taxi Pandas Profiling Report')
   
# save the report as html file
profile.to_file(output_file="taxi_profiling1.html")
   
# save the report as json file
profile.to_file(output_file="taxi_profiling2.json")
```

# Data quality issues

1. Check that total amount spent on a ride is bigger than 0. Cause: Sometimes negative fare amounts are stored in a database to show for amount refunded due to a variety of reasons e.g. incorrect charge on the card, cancelled, etc.

```python
df.loc[df['total_amount'] < 0]
```

2. Check that total_amount is calculated correctly

```python
sum_values = df['fare_amount']+ df['extra'] + df['mta_tax'] + df['tip_amount'] + df['tolls_amount'] + df['improvement_surcharge'] + df['congestion_surcharge']

df.loc[df['total_amount'] > sum_values]
```

3. Check that dropoff time is later than pickup

```python
df.loc[df['tpep_dropoff_datetime'] < df['tpep_pickup_datetime']]
```

# Descriptive statistic

Statistic for total_amount column (numerical data)

```python
df['total_amount'].describe()
```

```python
distance = df[df.trip_distance!= 0]
total = df['total_amount'].sum()
price = df[df.total_amount > 0]
```

```python
print("Total distance:", len(distance))
print("Total amount spent on taxi rides:", total)
print("The most expensive taxi ride:", price["total_amount"].max())
print("The most cheapest taxi ride:", price["total_amount"].min())
print("The shortest ride:", (distance["trip_distance"].min()*1.6), "km")
print("The longest ride:", (distance["trip_distance"].max()*1.6), "km")
```
